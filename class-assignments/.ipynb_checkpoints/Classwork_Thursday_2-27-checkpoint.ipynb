```python
import csv
import re
```

```python
inputpath = "bee_list.txt"
```

```python
authors = []
with open(inputpath, "r") as dset:
    data = csv.DictReader(dset, delimiter= "\t")
    for row in data:
        authors.append(row["Taxon Author"])
        
    header = data.fieldnames
print(header)
```

```python
# making string of authors
authstr = ''.join(authors)
# extracting all names
authreg = re.findall(r'\b[A-Za-z]+-?[A-Za-z]+\b', authstr)
authlist = []
# putting back into list
for i in authreg:
    authlist.append(i)
# new dict
authcount = dict()
# counting occurances of authors
for ii in authlist:
    authcount[ii] = authcount.get(ii, 0) + 1
# making sorted dict
from collections import OrderedDict
authsort = {key: val for key, val in sorted(authcount.items(), key=lambda item: item[1])}
```
